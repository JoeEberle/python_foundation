## Building Block #6  - Creating Python Functions

```python
def get_sentence_length(sentence): 
    return len(sentence) 

testing_function = True 
if testing_function:
    sentence = "Be happy." 
    print(get_sentence_length(sentence))
    sentence = "The cow is happy." 
    print(f"The sentence:{sentence} is {get_sentence_length(sentence) } characters long.") 
```

```python
import datetime
def what_day_is_it():
    # Get the current date and format it to get the day name
    current_day = datetime.datetime.now().strftime("%A")
    return current_day

# Example usage:
current_day = what_day_is_it()
print(f"Today is: {current_day} " ) 
```
